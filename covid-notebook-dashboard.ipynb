```python
#Â Import libraries
import requests
import pandas as pd
import plotly.graph_objects as go
from IPython.display import display
```

```python
# URL Link construction
timeline_url = "https://corona-api.com/timeline"
```

```python
# API call
raw_timeline = requests.request("GET", timeline_url)
raw_timeline = raw_timeline.json()
```

```python
# Transform data into pandas dataframe
df_timeline = pd.DataFrame(raw_timeline['data'])
df_timeline['updated_at'] = df_timeline['updated_at'].astype('datetime64')
df_timeline['date'] = df_timeline['date'].astype('datetime64')
```

```python
display(df_timeline.info())
display(df_timeline.head())
```

```python
# Show basic information from the data frame
summary = df_timeline.describe()
summary = summary.transpose()
summary.head()
```

```python
# Scatter plot for 3 variables
fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=df_timeline['date'], y=df_timeline['deaths'],
                    mode='markers',
                    name='deaths'))

fig.add_trace(go.Scatter(x=df_timeline['date'], y=df_timeline['recovered'],
                    mode='lines+markers',
                    name='recovered'))

fig.add_trace(go.Scatter(x=df_timeline['date'], y=df_timeline['active'],
                    mode='lines',
                    name='active'))

fig.show()
```

```python
# Scatter plot for 3 variables
fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=df_timeline['date'], y=df_timeline['new_deaths'],
                    mode='markers',
                    name='new deaths'))

fig.add_trace(go.Scatter(x=df_timeline['date'], y=df_timeline['new_recovered'],
                    mode='lines+markers',
                    name='new recovered'))

fig.add_trace(go.Scatter(x=df_timeline['date'], y=df_timeline['new_confirmed'],
                    mode='lines',
                    name='new confirmed'))

fig.show()
```
